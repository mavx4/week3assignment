```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
```

```python
df.head(10)
```

Removing Boroughs that are not assigned (1)

```python
df.drop(df[df['Borough']=='Not assigned'].index, inplace = True)
df.head(10)
```

Consolidating neighborhood with identical postcodes

```python
df_group=df.groupby('Postcode', as_index=False).agg(lambda x: ', '.join(set(x.dropna())))
df_group.head()
```

Assigning Borough name to unassigned Neighborhood

```python
df_group.Neighborhood = df_group.Borough.where(df_group.Neighborhood == 'Not assigned', df_group.Neighborhood)
```
